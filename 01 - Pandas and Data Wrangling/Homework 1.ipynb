## Task 2. RNA Sequences

In the `DATA_FOLDER/microbiome` subdirectory, there are 9 spreadsheets of microbiome data that was acquired from high-throughput RNA sequencing procedures, along with a 10<sup>th</sup> file that describes the content of each. 

Use pandas to import the first 9 spreadsheets into a single `DataFrame`.
Then, add the metadata information from the 10<sup>th</sup> spreadsheet as columns in the combined `DataFrame`.
Make sure that the final `DataFrame` has a unique index and all the `NaN` values have been replaced by the tag `unknown`.

### Solution

First, we need to open and read from each file of data. we set the "header" parameter to None because there are no header in those files. And we group them into a list of frames to be able to iterate on them.

```python
# Write your answer here

#open the 9 files
mb1 = pd.read_excel(DATA_FOLDER+'/microbiome/MID1.xls', sheetname='Sheet 1', header=None)
mb2 = pd.read_excel(DATA_FOLDER+'/microbiome/MID2.xls', sheetname='Sheet 1', header=None)
mb3 = pd.read_excel(DATA_FOLDER+'/microbiome/MID3.xls', sheetname='Sheet 1', header=None)
mb4 = pd.read_excel(DATA_FOLDER+'/microbiome/MID4.xls', sheetname='Sheet 1', header=None)
mb5 = pd.read_excel(DATA_FOLDER+'/microbiome/MID5.xls', sheetname='Sheet 1', header=None)
mb6 = pd.read_excel(DATA_FOLDER+'/microbiome/MID6.xls', sheetname='Sheet 1', header=None)
mb7 = pd.read_excel(DATA_FOLDER+'/microbiome/MID7.xls', sheetname='Sheet 1', header=None)
mb8 = pd.read_excel(DATA_FOLDER+'/microbiome/MID8.xls', sheetname='Sheet 1', header=None)
mb9 = pd.read_excel(DATA_FOLDER+'/microbiome/MID9.xls', sheetname='Sheet 1', header=None)

frames = [mb1,mb2,mb3,mb4,mb5,mb6,mb7,mb8,mb9]
```

Then, for each frame, we add a column with the index "BARCODE" to be able to recognize which lines belong to which file we read and to merge it easily with the metadata which has a column "BARCODE" that rely on the data files

```python
i = 1
for f in frames:
    f['BARCODE'] = 'MID'+str(i)
    i+=1
```

We can then concatenate the 9 DataFrame to get only one big resulting DataFrame. We have to set the "ignore_index" parameter to True to avoid duplicate indexes.

```python
mb = pd.concat(frames, ignore_index=True)
```

After that we can read the metadata from the metadata.xls file and store it into a DataFrame

```python
metadata = pd.read_excel(DATA_FOLDER+'/microbiome/metadata.xls', header=0)
```

And the last thing we need to do is to merge the two DataFrames on the column "BARCODE" to get one final DataFrame.

We also need to replace the NaN values by "unknown"

```python
result = pd.merge(mb, metadata, on='BARCODE').fillna('unknown')
result
```

If we check the uniqueness of the index, we can see that they are unique!

```python
result.index.is_unique
```

If we want to drop the "BARCODE" column we also can

```python
result_without_Barcode = result.drop('BARCODE', axis=1)
result_without_Barcode
```
